```python
import pandas as pd
import plotly.express as px
import plotly.io as pio

pio.renderers.default = "notebook_connected"

# Load dataset
df = pd.read_csv("nifty50.csv")


df_subset = df.iloc[50:101]


df_subset.columns = df_subset.columns.str.strip()


df_subset[df_subset.columns[1]] = pd.to_numeric(
    df_subset[df_subset.columns[1]],
    errors='coerce'
)


fig = px.bar(
    df_subset,
    x=df_subset.columns[0],
    y=df_subset.columns[1],
    title="Interactive Bar Chart (Rows 50 to 100)"
)

fig.show()
```

```python
import pandas as pd
import plotly.express as px
import plotly.io as pio

pio.renderers.default = "notebook_connected"
# Load dataset
df = pd.read_csv("nifty50.csv")
df_large = df.iloc[500:1001]
df_large = df_large.sort_values("dt")
y_col = df_large.select_dtypes(include="number").columns[0]
fig = px.line(
    df_large,
    x="dt",
    y=y_col,
    title="Line Chart (Rows 500–1000)"
)
fig.update_layout(
    xaxis=dict(rangeslider=dict(visible=True))
)

fig.show()
```

```python

df_numeric = df.select_dtypes(include="number")
Q1 = df_numeric.quantile(0.25)
Q3 = df_numeric.quantile(0.75)
IQR = Q3 - Q1
outliers = ((df_numeric < (Q1 - 1.5 * IQR)) | 
            (df_numeric > (Q3 + 1.5 * IQR))).sum()

print("Outliers in each column:\n")
print(outliers)
```
